### imports and graphics configurations ###

```python

import folium
from IPython.display import display
```

```python

from selenium import webdriver
import os
import time
```

```python
#webdriver constants
delay=5
fname="testmap.html"
fileURL="file://{path}/{mapfile}"
boundaryStyleDict={'color':'red'}
```

```python
import pandas as pd
edge_lane = pd.read_pickle('../data/edge_lane_w_coords.pkl')
```

```python
ATHENS_center = ( 37.9829088,23.7332314)

myMap = folium.Map(location=ATHENS_center, zoom_start=100.5,tiles ='Stamen Toner',zoom_control=False)

for row in edge_lane.iterrows():
    road_seg = {"coordinates": [(row[1]['node_1_lon'],row[1]['node_1_lat']), (row[1]['node_2_lon'],row[1]['node_2_lat'])], "type": "LineString"}
    
    road=folium.GeoJson(data=road_seg,style_function=lambda x: boundaryStyleDict)
    road.add_to(myMap)

display(myMap)
```

```python
myMap.save(fname)

browser = webdriver.Chrome()
browser.get(fileURL.format(path=os.getcwd(),mapfile=fname))
#Give the map tiles some time to load
time.sleep(delay)
browser.save_screenshot('Athens.png')
browser.quit()
```
