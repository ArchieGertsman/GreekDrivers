# Visualizing trajectory data

```python
import pandas as pd
import random
import pydeck as pdk
import json
```

##### Here we would be visualizing the data for block 4 between 8:30 - 9:00 on 24/10/2018

```python
df = pd.read_pickle("block4_1.pkl")
df.head()
```

```python
#processing trajectories
df = df.reset_index()
coordinates = [[k,[[i, j] for i,j in zip(df[df['id'] == k]['lon'], df[df['id'] == k]['lat'])],list(df[df['id'] == k].time)] for k in df.id.unique()]
id,coordinate_list,timestamp = zip(*coordinates)
dfx = pd.DataFrame({"id" : id,"coordinates": coordinate_list, "timestamps" : timestamp})
dfx.head()
```

```python
#generating unique colour for each trajectory
X=np.random.randint(10,20)
dfx['color'] = [(np.random.randint(1,254),np.random.randint(1,254),np.random.randint(1,254)) for i in dfx.id.unique()]
```

```python
#defining layer object for pydeck
layer = pdk.Layer(
    "TripsLayer",
    dfx,
    get_path="coordinates",
    get_timestamps="timestamp",
    get_color="color",
    opacity=0.5,
    width_min_pixels=2,
    rounded=True,
    trail_length=600,
    current_time=50,
) 
view_state = pdk.ViewState(latitude=37.981384, longitude=23.734019, zoom=17, bearing=0, pitch=45)
```

```python
#visualizing the dataset
r = pdk.Deck(layers=layer, initial_view_state=view_state)
r.to_html("block_4_trajectories.html")
```
