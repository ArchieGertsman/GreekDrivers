### imports and graphics configurations ###

```python
import numpy
import pandas
import geojson
```

### constants ###

```python
in_fname="../Data/sample_larger.csv"
out_fname="./Athens"
colorDict={
    "Car":[0,0,255],#"blue",
    "Motorcycle":[0,255,255],#"cyan",
    "Bus":[0.255,0],#"green",
    "Taxi":[255,255,255],#"white",
    "Medium Vehicle":[238,130,238],#"violet",
    "Heavy Vehicle":[173, 216, 230],#"lightblue"
}
```

### code starts here###

```python
with open(in_fname, "r") as f:
    temp=f.readlines()
    
(header,data)=(temp[0],temp[1:])
```

```python
print(header)
```

```python
print(data[:4])
```

```python
def process(in_str):
    try:
        in_str=in_str.strip().strip(";")
        data_len=6
        parts=in_str.split(";")
        (header,temp)=(parts[:4],numpy.array(parts[4:],dtype=numpy.float))
        properties={"track_id":header[0].strip(),"type":header[1].strip(),"traveled_d":float(header[2]),"avg_speed":float(header[3])}
        #print(properties["type"])
        properties["color"]=colorDict[properties["type"]]
        temp=temp.reshape(-1,data_len)[:,[1,0]]
        temp=geojson.LineString(temp.tolist())
        return geojson.Feature(geometry=temp,properties=properties)
    except Exception:
        print("error; properties:"+str(properties))
        return None
    
    
process(data[1])
```

```python
LINEDATA=geojson.dumps(geojson.FeatureCollection([process(line) for line in data]))
```

```python
#print(LINEDATA)
```

```python
outscript="var linedata="+LINEDATA
    
with open(out_fname+".js","w") as f:
    f.write("var linedata="+LINEDATA)
    
with open(out_fname+".json","w") as f:
    f.write(LINEDATA)
```
