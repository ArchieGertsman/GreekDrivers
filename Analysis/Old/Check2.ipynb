```python
import numpy as np
import nvector as nv
import pandas as pd
import osmnx as ox
import osmapi as osm
import networkx as nx
import matplotlib.pyplot as plt


def cross_track_dist(latA1,lonA1,latA2,lonA2,latB,lonB):
    frame = nv.FrameE(a=6371e3, f=0)
    
    
    
    
    
    pointA1 = frame.GeoPoint(latA1, lonA1, degrees=True)
    pointA2 = frame.GeoPoint(latA2, lonA2, degrees=True)
    pointB = frame.GeoPoint(latB, lonB, degrees=True)
    pathA = nv.GeoPath(pointA1, pointA2)

    s_xt = pathA.cross_track_distance(pointB, method='greatcircle')
    d_xt = pathA.cross_track_distance(pointB, method='euclidean')
    
    if(s_xt<0):
        s_xt=-1*s_xt
    
    val_txt = '{:4.2f}'.format(s_xt/1000)
    #'Cross track distance: s_xt, d_xt = {}'.format(val_txt)
    return val_txt
```

```python
datafile="/home/shekhar/Desktop/UIUC_Project3/Analysis/Test_Pairs.csv"
df = pd.read_csv(datafile,sep=',')
#df.dropna(how='all', axis='columns')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
#print(df.head(10))
a,b=float(df.iloc[0][1]),float(df.iloc[0][2])
#print(a,b)
df=df.drop('760',axis=1)
df=df.replace([' '],[np.nan])
print(df.iloc[0][2])
```

```python
i=0
j=1
G = ox.graph_from_point((a,b), network_type='drive', dist=800, simplify=True)

while i < 9:
    j=1
    while j < 252:
        if(df.iloc[i][j]!=np.nan):
            latB=float(df.iloc[i][j])
            lonB=float(df.iloc[i][j+1])
            #print(latB,lonB)
            surrounding_nodes = ox.get_nearest_edge(G,(latB,lonB))
            latA1= G.nodes[surrounding_nodes[0]]['y']
            lonA1= G.nodes[surrounding_nodes[0]]['x']
            latA2= G.nodes[surrounding_nodes[1]]['y']
            lonA2= G.nodes[surrounding_nodes[1]]['x']
            df.iat[i,j]=float(cross_track_dist(latA1,lonA1,latA2,lonA2,latB,lonB))
            #print(df.iat[i,j])
        j=j+2
    i=i+1        
#df=df.replace(['nan,nan',' ,nan'],[np.nan,np.nan])
df.head(10)



#df.dropna(axis='columns',how='all')
#df.head()
#fig, ax = ox.plot_graph(G, show=False, close=False)
#plt.show()
#plt.savefig('./sample_map.png')
#fig.savefig('/home/shekhar/Desktop/UIUC_Project3/Analysis/sample_map.png')
```
