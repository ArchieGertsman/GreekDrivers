```python
import pandas as pd
import matplotlib.pyplot as plt
```

### comparing trajectories with different speed cutoffs

```python
df = pd.read_pickle("acc_75p_0_lane_filter_f1.pkl")
```

#### F1 Score: trajectories with speed > 0 75% of time 

```python
df1 = df.loc[(slice(None),['f1_score','traj_len'],slice(None))]
df1
```

```python
fig,ax = plt.subplots(2)
df1 = df.loc[(slice(None),'f1_score',slice(None))]
fig.set_size_inches(8,10)
fig.set_dpi(80)
fig.tight_layout(h_pad=3)
fig.suptitle('F1_score vs. Trajectory Length')
plt.subplots_adjust(top=0.90)
model_keys = df1.index.get_level_values(0).unique()[0:4]

ax_mean, ax_std = ax

ax_mean.set_ylabel('Mean')
ax_std.set_ylabel('Std')

for i,name in enumerate(model_keys):
    ax_mean.plot(df1.columns, df1.loc[(name, 'mean')])
    ax_std.plot(df1.columns, df1.loc[(name, 'std')])

fig.legend(model_keys)
plt.show()
```
